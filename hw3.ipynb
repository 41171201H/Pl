```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# 發送 HTTP 請求並獲取網頁內容
url = "https://tw.stock.yahoo.com/cryptocurrencies"
response = requests.get(url)

# 檢查響應狀態碼，確保獲取網頁成功
if response.status_code == 200:
    page_content = response.content

    # 使用 BeautifulSoup 解析 HTML
    soup = BeautifulSoup(page_content, "html.parser")

        
# 找到指定 class 的 <ul> 標籤
ul_tag = soup.find('ul', class_='M(0) P(0) List(n)')


# 找到所有指定 class 的 <li> 標籤
li_tags = ul_tag.find_all('li', class_='List(n)')


# 初始化存放資料的列表
stock_names = []
stock_codes = []
stock_price = []



# 遍歷每個 <li> 標籤，提取其文本
for li_tag in li_tags:
   # 找到每個<li>中的所有<span>元素
   
    #股名
    divs = li_tag.find_all('div', class_="Lh(20px) Fw(600) Fz(16px) Ell")
    for div in divs:
        stock_names.append(div.get_text())
        
    #股號
    spans = li_tag.find_all('span', class_='Fz(14px) C(#979ba7) Ell')
    for span in spans:
        stock_codes.append(span.get_text())
        
    #股價
    spans = li_tag.find_all('span', class_=['Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-up)', 'Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-down)'])
    for span in spans:
        stock_price.append(span.get_text())
        

print(len(stock_names))
print(len(stock_codes))
print(len(stock_price))

# 將資料轉換成 DataFrame
df = pd.DataFrame({
    '股名': stock_names,
    '股號': stock_codes,
    '股價': stock_price,
    
})

# 設置 Pandas 顯示選項，將顯示的寬度設置為無限制
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# 印出 DataFrame
print(df)

# 將DataFrame轉換為字典
data_dict = df.to_dict(orient='records')
# 打印字典
print(data_dict)

# 將 DataFrame 輸出為 CSV 文件
df.to_csv('hw3.csv', index=False)
# 將 DataFrame 輸出為 JSON 文件
df.to_json('hw3.json', orient='records')
```
